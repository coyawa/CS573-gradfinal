```python
import re
fhand = open("election data.txt", "r")
filelist = []
s =''
for i in fhand:
    s += i
    filelist.append(i)
statelist = re.findall("([A-Z].+\t\n)", s)

statelist
```

```python
statedatalist = []
indextemp = []
for i in statelist:
    indextemp.append(filelist.index(i))
for i in range(len(indextemp)):
    if i == 0:
        statedatalist.append(filelist[:indextemp[i]])
        statedatalist.append(filelist[indextemp[i]:indextemp[i+1]])
    elif i == len(indextemp)-1:
        statedatalist.append(filelist[indextemp[i]:])
    else:
        statedatalist.append(filelist[indextemp[i]:indextemp[i+1]])
statedatalist
```

```python
for i in range(len(statedatalist)):
    for j in range(len(statedatalist[i])):
        statedatalist[i][j] = statedatalist[i][j].rstrip("\n").rstrip("\t").lstrip("\t")
    if i == 0:
        temp1 = statedatalist[i][0].split("\t")
        statedatalist[i] = temp1
    else:
        temp1 = statedatalist[i][2].split("\t")
        temp2 = []
        for m in range(len(statedatalist[i])):
            if m !=2 :
                temp2.append(statedatalist[i][m])
            else:
                for n in temp1:
                    temp2.append(n)
        statedatalist[i] = temp2
    
    
    
statedatalist
```

```python
import csv
with open('election.csv', 'w') as csvfile:
    writer = csv.writer(csvfile, lineterminator='\n')
    writer.writerows(statedatalist)
```

```python
title = ["STATES"]
for i in statedatalist[0]:
    title.append(i)
title
```

```python
jsondata = []

statedatalist = statedatalist[1:]
for i in range(len(statedatalist)):
    datadict = {}
    for j in range(len(statedatalist[i])):
        datadict[title[j]] = statedatalist[i][j]
    jsondata.append(datadict)
jsondata
```

```python
import json
with open('election.json', 'w') as outfile:
    json.dump(jsondata, outfile)
```
